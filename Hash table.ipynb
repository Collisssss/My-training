```python
class Solution:
    def uncommonFromSentence(self, A, B):
        from conllections import Counter
        count = Counter((A+''+B).split())
        return [word for word ,c in count.item() if c==1]
```

list(set([])) 去重复在转化成array

而且 set是哈希表结构

第二种解法

```python
def containsDuplicate(nums):
    # 创建一个空集合
    seen = set()

    # 遍历每一个元素
    for num in nums:
        # 如果该元素在集合中已经存在，说明存在重复的元素
        if num in seen:
            return True
        # 否则将这个元素添加到集合中
        seen.add(num)

    # 如果循环结束还没有返回，说明没有重复的元素
    return False

# 测试
nums1 = [1, 2, 3, 4, 5]
nums2 = [1, 2, 3, 4, 5, 2]

print(containsDuplicate(nums1))  # 输出: False
print(containsDuplicate(nums2))  # 输出: True
```
