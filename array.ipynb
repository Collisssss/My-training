```python
def twoSum(self, nums: list[int], target: int) -> list[int]:
    h = {}
    for idx, val in enumerate(nums):
        n = target - val
        if n not in h:
                   h[val] = idx
        else:
                   return [h[n], idx]
```

如果保留空格大小不保留反转呢

```python
import re

def reverse_words_preserve_space(s):
    # Split the string into a list of words and spaces
    segments = re.split(r'( +)', s) # re.split() r表示原始字符串， ‘（ +）’为提取这个部分

    # Reverse the order of the words, but not the spaces
    segments[::2] = segments[::2][::-1]

    # Join the words and spaces back into a string
    return ''.join(segments)
```

```python
print(reverse_words_preserve_space(s))
```

import re 这条语句是在Python程序中导入 re 模块，这样就可以在程序中使用 re 模块提供的函数和类了。
re 模块提供了很多函数，如 match、search、findall、split、sub 等，用于进行各种复杂的字符串操作。
例如，re.split(pattern, string) 函数可以根据 pattern 参数指定的正则表达式，将 string 参数指定的字符串进行分割。
在正则表达式中，空格字符 ' ' 用于匹配空格，+ 是一个特殊字符，表示前面的字符（在这个例子中是空格）可以出现一次或多次。

所以 ' ' 表示一个空格，' ' 表示两个连续空格，以此类推。但如果我们要匹配一个或多个连续空格，不论空格的数量是多少，我们可以使用 ' +' 这个正则表达式。

这就是 ' +' 用于匹配一个或多个连续空格的含义。这个正则表达式可以匹配任意数量（至少一次）的连续空格。

在正则表达式中，( +)表示匹配一个或多个连续的空格。其中的圆括号用于捕获匹配的文本，这样在使用re.split函数分割字符串时，匹配的文本（也就是空格）也会被包含在结果列表中。

在re.split(r'( +)', s)中，r'( +)'是一个正则表达式，用于匹配一个或多个连续的空格。re.split函数会将字符串s在每个匹配的位置进行分割，并返回一个包含了分割出的片段的列表。由于正则表达式中包含了圆括号，所以分割出的片段中也会包含匹配的空格。








如果还要保留空格大小还反转呢

```python
import re 
def reverse_words_nopreserve_space(s):
    segment = re.split(r'( +)', s)
    segment = segment[::-1]
    return ''.join(segment)
```

```python
print(reverse_words_nopreserve_space(s))
```

```python
class Solution:
    def letterCombinations(self, digits: str) -> list:
        if digits == "":
            return []

        letter_map = {
            '2': ['a', 'b', 'c'],
            '3': ['d', 'e', 'f'],
            '4': ['g', 'h', 'i'],
            '5': ['j', 'k', 'l'],
            '6': ['m', 'n', 'o'],
            '7': ['p', 'q', 'r', 's'],
            '8': ['t', 'u', 'v'],
            '9': ['w', 'x', 'y', 'z']
        }

        result = ['']
        for digit in digits:
            temp_list = []
            for combo in result:
                for letter in letter_map[digit]:
                    temp_list.append(combo + letter)
            result = temp_list

        return result
```

```python
x = "46"
sol = Solution()
sol.letterCombinations(x)
```
